```python
import pandas as pd

# Load your file
file_path = 'output_cleaned.csv'  # update with your actual filename
df = pd.read_csv(file_path)

# Remove rows where voter_id is NaN (null)
initial_len = len(df)
df = df.dropna(subset=['voter_id'])

# Also remove rows where voter_id might be just an empty string or whitespace
df = df[df['voter_id'].astype(str).str.strip() != ""]

# Save back to the same file
df.to_csv(file_path, index=False, encoding='utf-8-sig')

print(f"Rows removed: {initial_len - len(df)}")
print("File updated successfully.")
```

```python
# Check if the 'ward' column exists and print unique values
if 'ward' in df.columns:
    unique_wards = df['ward'].unique()
    print("Unique Ward Values:")
    print("-" * 20)
    for ward in unique_wards:
        print(ward)
else:
    print("Column 'ward' not found in the dataset.")
```

```python
# Replace 'वडा न॰ ०' with 'वडा न॰ १'
if 'ward' in df.columns:
    df['ward'] = df['ward'].replace('वडा न॰ ०', 'वडा न॰ १')
    
    # Save the updated data back to the same file
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
    print("Updated 'वडा न॰ ०' to 'वडा न॰ १' successfully.")
else:
    print("Column 'ward' not found.")
```

```python
# Replace 'वडा न॰ ०' with 'वडा न॰ ४'
if 'ward' in df.columns:
    # We replace the specific string 'वडा न॰ ०' with 'वडा न॰ ४'
    df['ward'] = df['ward'].replace('वडा न॰ ~०', 'वडा न॰ ४')
    
    # Save the updated data back to the same file
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
    print("Updated 'वडा न॰ ०' to 'वडा न॰ ४' successfully.")
    
    # Verify the change by printing unique values again
    print("Current Unique Wards:", df['ward'].unique())
else:
    print("Column 'ward' not found.")
```

```python
# Replace 'वडा न॰ ०' with 'वडा न॰ ९'
if 'ward' in df.columns:
    # Update the values in the 'ward' column
    df['ward'] = df['ward'].replace('वडा न. ०', 'वडा न॰ ९')
    
    # Save back to the same CSV file
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
    
    print("Updated 'वडा न॰ ०' to 'वडा न॰ ९' successfully.")
    print("Updated Wards:", df['ward'].unique())
else:
    print("Column 'ward' not found.")
```
